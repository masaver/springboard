```python
import omdb
import json
import numpy as np
import pandas as pd
import time
from IPython.display import clear_output

omdb.set_default('apikey', 'f022982e')
```

```python
#Read csv with names and IDs of movies, 
df_movies = pd.read_csv('movie_titles.csv')
df_movies.index.name = 'MovieID'
df_movies.index = np.arange(1, len(df_movies)+1)
df_movies['Is_there_content'] = np.nan
df_movies['Actors'] = None
df_movies['Genre'] = None
df_movies['Country'] = None
df_movies['Director'] = None
df_movies['Runtime'] = None
df_movies['Writer'] = None
df_movies['Movie']
df_movies.head(10)
```

```python
for j in range(0,len(df_movies.index)):
    movie = omdb.title(df_movies.iloc[j,0], timeout = 100)
    if movie != {}:
        df_movies.iloc[j,3] = 'YES'
        df_movies.iloc[j,[4,5,6,7,8,9]] = [movie.actors, movie.genre, movie.country, movie.director, movie.runtime, movie.writer]
    else:
        df_movies.iloc[j,3] = 'NO'
```
