```python
#%matplotlib inline
#import ast
#import json
#import time
#from IPython.display import clear_output
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt
#plt.style.use('seaborn')
#print(plt.style.available)
```

```python
#Read csv with names and IDs of movies, 
df_movies = pd.read_csv('movies_with_info.csv')
df_movies.index.name = 'MovieID'
df_movies.index = np.arange(1, len(df_movies)+1)
df_movies_2 = pd.read_pickle('movies_plus_info.pkl')
df_movies['Genres'] = df_movies_2['Genre']
df_movies.head(10)
```

```python
genre_list = []

for x in df_movies.Genres.values:
    if x != None:
        y = x.split(',')
        for genre in y:
            if genre != genre_list:
                genre_list.append(genre.replace(' ',''))
                
genre_list = list(set(genre_list))
print(genre_list)
```

```python
#This part  creates a dictionary with movie IDs as keys, and movie genre as values
a = df_movies.index.values
b = df_movies['Genres'].values
d = zip(a,b)
d = dict(d)

#Creates a dictionary with Movie IDs as keys, and Year of movie release as values
movie_dict = dict(zip(df_movies.index, df_movies.Year))

#Also, Creates a second dictionary with Movie IDs as keys, and Movie names as values
name_dict = dict(zip(df_movies.index, df_movies.Name))
```

```python
#Read the data frame with all data, and convert dates to a datetime object
#It also creates a new column, 'Year of Release', with the years in which the respective movies where released.
#Finally,the first 5 rows of the data frame are shown.
df_data = pd.read_csv('all_data_2.csv')
df_data['Date'] = pd.to_datetime(df_data['Date'], format='%Y-%m-%d %H:%M:%S.%f')
df_data['Year of Release'] = np.nan
df_data['Genre'] = np.nan
df_data['Genre'] = df_data['Movie ID'].map(d)
df_data['Year of Release'] = df_data['Movie ID'].map(movie_dict)

df_data.head(5)
```

```python
for element in genre_list:
    df_movies[element] = np.nan

df_movies.head()
```

```python
dum = lambda x: x if x != None else 'None'
df_movies['Genres'] = df_movies['Genres'].apply(dum)
```

```python
for elem in genre_list:
    print(elem)
    
    dummy = lambda x: 'Y' if elem in [x.replace(' ','') for x in x.split(',')] else 'N'
    df_movies[elem] = df_movies['Genres'].apply(dummy) 
```
